```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.grid_2d_graph(m=5, n=5)
pos = nx.nx_pydot.graphviz_layout(G, prog='dot')

plt.figure(figsize=(5, 5))
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
plt.axis('off')
plt.savefig('e1.png')
```
